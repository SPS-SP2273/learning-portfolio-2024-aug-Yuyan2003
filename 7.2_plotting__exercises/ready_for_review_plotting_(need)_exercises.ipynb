## Exercise 1 (Fill in the blanks)

```python
from matplotlib import pyplot as plt

# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color="brown", linestyle="dashed",
             marker="p", markerfacecolor="yellow", markeredgecolor="brown",
             ecolor="black",capsize=5,
             label="Y values with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

## Exercise 2 (A simple mathematical plot)  

```python
import numpy as np

x = np.linspace(0, 50, 1000)
#x = np.array(range(51))
y = (np.exp((-x)/10))*np.sin(x)

plt.plot(x, y, label='sin(x)',color='blue', )
plt.xlabel('x', color='darkorange', size=15)
plt.ylabel('y', color='darkorange', size=15, rotation=0) #Rotate y-axis label to upright position
plt.title('y=$e^{-x/10}sin(x) $', size=20) #To display multi-letter subscripts or superscripts correctly, you should put them in curly braces 
plt.grid(alpha=.25)
plt.ylim([-1,1])

plt.show()
```
