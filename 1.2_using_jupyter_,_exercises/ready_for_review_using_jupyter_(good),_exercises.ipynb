```python
squares = [i**2 for i in range(10)]

print(squares)
```

```python
%%timeit

squares = [i**2 for i in range(10)]

#print(squares)
```

```python
%%timeit

squares = list()
for i in range(10):
    squares.append(i**2)

#print(squares)
```
