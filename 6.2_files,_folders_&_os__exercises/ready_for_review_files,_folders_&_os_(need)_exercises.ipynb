```python
#Task 2

def rename_my_file(old_file_name):
    dictionary= {'jan': '01', 'feb': '02', 'mar': '03',
             'apr': '04', 'may': '05', 'jun': '06',
             'jul': '07', 'aug': '08', 'sep': '09',
             'oct': '10', 'nov': '11', 'dec': '12'}

    for key in dictionary.keys():
        if key in old_file_name:
            new_file_name = old_file_name.replace(key, dictionary[key])
            #print(new_file_name.replace("_","-"))
            return new_file_name.replace("_","-") #able to store the output

old_file_name = "oct-08_singapore.txt"
new_file_name = rename_my_file(old_file_name)
print(new_file_name)
```

```python
#Task 4

import os
import glob

dir = "os-collaboration-exercise-data"

path = os.path.join(dir, '*.txt')
x = glob.glob(path)

for y in x:  
    filename = y.split(os.path.sep) [-1]
    #print(f'{filename} ----> {rename_my_file(filename)} ')
    new_file_name = rename_my_file(filename)
    new_path = os.path.join(dir, new_file_name)
    print(f'{y} has been renamed to {new_path}')
    os.rename(y, new_path)
```

```python
#Task 6

dir = "os-collaboration-exercise-data"
path = os.path.join(dir, '*cairo*')
cairo_files = glob.glob(path)
#print(cairo_files)

import shutil

cairo_destination = os.path.join('cities', 'Cairo')

# Create the Cairo destination directory if it does not exist
try:
    os.mkdir(cairo_destination)
    print(f'Creating {cairo_destination}')
except FileExistsError:
    print(f'{cairo_destination} already exists; skipping creation.')

#Copy files that match "cairo" to the 'cities/Cairo' folder
for file in cairo_files:
    shutil.copy(file, cairo_destination)
    print(f'Copied {file} to {cairo_destination}')
```

```python
#Task 7

dir = "os-collaboration-exercise-data"

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila', 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

import fnmatch

all_files = glob.glob(os.path.join(dir, '*'))

for city in cities:
    city_pattern = f'*{city.lower()}*' #Create the pattern to match files related to the current city (case insensitive)
    
    city_files = [file for file in all_files if fnmatch.fnmatch(file.lower(), city_pattern)] #Filter files that match the current city pattern (case insensitive)
    
    city_destination = os.path.join('cities', city)

#check for existence of this file. do not create if it exists already.
    try:
        os.mkdir(city_destination)
        print(f'Creating {city_destination}')
    except FileExistsError:
            print(f'{city_destination} already exists; skipping creation.')

    for file in city_files:
        shutil.copy(file, city_destination)
        print(f'Copied {file} to {city_destination}')
```
