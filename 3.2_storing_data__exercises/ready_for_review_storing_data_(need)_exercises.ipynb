## Exercise 5 (A Matter of Statistics)

Equation of Pearson correlation coefficient (r) for reference



```python
x_1 = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
x_2 = [1, 8, 27, 64, 125, 216, 343, 512, 729, 1000]

print(f'The mean of X1 is {np.mean(x_1)}.')
print(f'The mean of X2 is {np.mean(x_2)}.')

print(f'The standard deviation of X1 is {np.std(x_1):.2f}.')
print(f'The standard deviation of X1 is {np.std(x_2):.2f}.')

#without using np.corrcoef() function
xi_xbar = x_1[0] - np.mean(x_1)
yi_ybar = x_2[0] - np.mean(x_2)

r_numerator = np.sum(xi_xbar * yi_ybar)
#print(r_numerator) #just to check whether the code works

r_denominator = np.sqrt(np.sum(xi_xbar**2) * np.sum(yi_ybar**2))
#print(r_denominator) #just to check whether the code works

print(f'The Pearson correlation coefficient for the two datasets is {r_numerator / r_denominator}.')

#using np.corrcoef() function
print(f'The Pearson correlation coefficient of X1 is {np.corrcoef(x_1)}.')
print(f'The Pearson correlation coefficient of X2 is {np.corrcoef(x_2)}.')
```
